# Workshop Setup

You will:
1. Check your Python version
2. Create a virtual environment (`.venv`)
3. Install all dependencies
4. Register a Jupyter kernel so you can select it in `train.ipynb`

> **IMPORTANT:** Shell commands here use Unix paths (`/`). If you're on **Windows**, replace `.venv/bin/` with `.venv\Scripts\`.

## Step 1 — Check Python version

Python 3.10 or higher is required.

```python
import sys

if (sys.version_info[0], sys.version_info[1]) < (3, 10):
    raise RuntimeError("Python 3.10+ is required. Raise your hand if you need help. (Don't be shy)")
else:
    print("✓ Version check passed")
```

## Step 2 — Create virtual environment

Creates a `.venv` folder in the project directory.

```python
!python -m venv .venv

print("✓ Virtual environment created at .venv/")
```

## Step 3 — Install dependencies

Installs libraries. This may take a minute or two.

```python
!.venv/bin/pip install --upgrade pip --quiet
!.venv/bin/pip install -r requirements.txt --quiet

print("✓ Dependencies installed")
```

## Step 4 — Register the Jupyter kernel

Makes the venv available as a selectable kernel inside Jupyter.

```python
!.venv/bin/python -m ipykernel install --user --name=wandb-demo --display-name="wandb-demo"

print("✓ Kernel registered")
```

## All done!
